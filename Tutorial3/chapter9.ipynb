## Extracting row and columns

```python
import pandas as pd
data = pd.read_csv("national-health-and-nutrition-examination-survey/questionnaire.csv")
```

```python
# Import only some columns (need to specify columns by their index number (starting at 0) using 'usecols')

data_some_cols = pd.read_csv("national-health-and-nutrition-examination-survey/questionnaire.csv", 
                            usecols = [0,1,2,3,4,5,6])

data_some_cols_2 = pd.read_csv("national-health-and-nutrition-examination-survey/questionnaire.csv",
                    usecols= ['SEQN','ACD011A', 'ACD011B', 'ACD011C', 'ACD040', 'ACD110', 'ALQ101'])
```

```python
print(data_some_cols)
```

```python
print(data_some_cols_2)
```

```python
# Refer to specific columns
print(data['SEQN'])
print(data.SEQN)
```

```python
# if we're interested in more than one column though, we have to use
print(data[['SEQN','ACD011A', 'ACD011B']])   # REMEMBER: lists are enclosed in [] !!! (-> double [])
```

### Filtering by Rows

```python
data_some_rows = data[1:4]
data_some_rows
```

```python
# Asking for a single row instead of a range? => getting an error! we need to specify a range
# data[2] would give an error
# we have to use:
data[0:1] # to get row 0
```

```python
# OR
data[:1]
```

```python
# When we return rows, the ':' is always required. You can use : by itself to return all of the rows
data[:]
```

### Using criteria to filter rows

```python
data_some_row = data[(data.ACD040 == 1.0)]
data_some_row
```

```python
data_some_other_rows = data[(data.ALQ101 == 1.0) & (data.ALQ120Q == 1.0)]
data_some_other_rows
```

```python
# combining row selection with column selection
some_rows = data[(data.ALQ101 == 1.0) & (data.ALQ120Q == 1.0)][['SEQN', 'ALQ101', 'ALQ120Q']]
some_rows
```

```python
# selecting rows on the row index 
some_rows = data.iloc[0:4]
some_rows
```

```python
# select the first row from the Dataframe
some_rows = data.iloc[0]
some_rows
```

```python
# using the range() function to indicate the records that we want
some_rows = data.iloc[range(0, len(data), 100)]
some_rows
```

```python
# Can also specify columns ranges using the iloc method again using the column index numbers
some_rows = data.iloc[range(0, len(data), 100), 0:4]
some_rows
```

```python
some_rows = data.iloc[range(0, len(data), 100), [0, 3, 7, 10, 12]]
some_rows
```

```python
# same, using columns names (using loc method!!!!)
some_rows = data.loc[range(0, len(data), 100), ['SEQN','ACD011A', 'ACD011B']]
some_rows
```

## Sampling

```python
# sample method, which allows you to extract a sample of the records from the Dataframe

data.sample(10, replace=False)  # 10 records, do not select the same record twice (this is the default)
```

```python
data.sample(frac=0.05, random_state=1) # 5% of records, same records if run again
```
